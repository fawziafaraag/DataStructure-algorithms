```python
class student:
    def __init__(self,id_,name,group):
        self.ID=id_
        self.NAME=name
        self.GROUP=group
        
    def __repr__(self):
        return f"id={self.ID}\nname = {self.NAME}\ngroup= {self.GROUP}"

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

    def __repr__(self):
        return str(self.data)
    
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def displayAll(self):
        nd=self.head
        print("ID\tName\tGroup")
        while(nd != None):
            print(nd.data)
            nd = nd.next
            
    def append(self,student):
        nd = Node(student)
        if self.head ==None:
            self.head = nd
            self.tail =nd
            self.size+=1
        else:
            nd.prev = self.tail
            self.tail.next = nd
            self.tail = nd
            self.size+=1
            
            
    def insertByPos(self,student, index):
        nd =Node(student)
        if self.head == None : # no node
            self.head = nd
            self.tail =nd
            self.size+=1          
        else: 
            if index == 0 :
                nd.next = self.head
                self.head.prev = nd
                self.head = nd
                self.size+=1
                
            else:
                i = 0
                cur = self.head
                while (i < index-1) & (cur!=None):  
                    cur = cur.next
                    i+=1                   
                if (cur == self.tail) | (cur==None) :
                    nd.prev = self.tail
                    self.tail.next = nd
                    self.tail = nd
                    self.size+=1
                else :
                    nd.next = cur.next 
                    cur.next.prev = nd
                    
                    nd.prev = cur
                    cur.next = nd
                    self.size+=1
                      
    def searchByName(self, name):
        nd = None 
        if self.head != None :
            nd = self.head
            while nd.data.name != name:
                nd = nd.next
                if(nd==None):
                    break   
        if nd:
            print("\nID\tName\tGroup\n",nd)
        else:
            print("\nNOT Found")
    def searchByName(self, NAME):
        nd = None 
        if self.head != None :
            nd = self.head
            while nd.data.NAME != NAME:
                nd = nd.next
                if(nd==None):
                    break 
        if nd:
            print("\nID\tName\tGroup\n",nd)
        else:
            print("\nnot found")
    
    def deleteById(self,ID):
        delete = 0 
        nd = None 
    
        if(self.head != None):
            nd = self.head

            if(self.head.data.ID == ID):
                if(self.head == self.tail): 
                    self.head = None 
                    self.tail = None
                    self.size-=1
                    delete = 1 
                    
                else: 
                    self.head = self.head.Next
                    self.head.prev = None
                    self.size-=1
                    delete = 1       
            else:
                while nd.data.ID != ID:
                    nd = nd.next
                    if(nd==None):
                        break
                if nd != None:
                    if nd == self.tail:
                        self.tail = self.tail.prev
                        self.tail.next = None 
                        self.size-=1
                        delete = 1
            
                    else: 
                        nd.prev.next = nd.next
                        nd.next.prev = nd.prev 
                        self.size-=1
                        delete = 1
                
        if delete:
            print("Done")
        else:
            print("Not Delete")

    def deleteAllByGroup(self,GROUP):
        nd = None 
        if self.head != None :
            nd = self.head
            while nd != None :
                if nd.data.GROUP == GROUP:
                    self.deleteById(nd.data.ID)
                nd = nd.next
                                       
    def deleteAll(self):
        self.head = None 
        self.tail = None 
        self.size=0
        print("all deleted!!")
           
    def __repr__(self):
        return f"\nList Size : {self.size}\nHead : {self.head} \nTail : {self.tail}"
        
```

# Stack task

```python
class Student:
    def __init__(self,i,n,g):
        self.Id = i
        self.name = n
        self.group = g
    def __repr__(self):
        return f"{self.Id}\t{self.name}\t{self.group}"
    
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        
    def __repr__(self):
        return str(self.data)
    
    
class StackLinkedList:
    def __init__(self):
        self.head = None 
        self.size = 0
        
    def displayAll(self):
        if self.head:
            nd=self.head
            while(nd != None):
                print(nd.data)
                nd = nd.prev
                
                
    def __repr__(self):
        return f"\nHead of stack : {self.head} \nstack size : {self.size}"

    
class Stack:
    def __init__(self, slist):
        self.slist = slist
        
    def push(self,student):
        nd= Node(student)
        if self.slist.head == None:
            self.slist.head = nd
            self.slist.size+=1
        else:
            nd.prev = self.slist.head
            self.slist.head = nd
            self.slist.size+=1
            
    def pop(self) :
        nd = None  
        if self.slist.head != None:
            nd = slist.head
            self.slist.head = self.slist.head.prev
            self.slist.size-=1
        
        return nd
    
    def __repr__(self):
        self.slist.displayAll()
        return str()   
    
    
```

```python
slist= StackLinkedList()
stk = Stack(slist)

s1=Student(1,"fawzia",'A')
s2=Student(2,"faraag",'B')
s3=Student(3,"mohamed",'C')
s4=Student(4,"wageeh",'B')
s5=Student(5,"marwa",'C')

# push
print("push")
stk.push(s1)
stk.push(s2)
stk.push(s3)
stk.push(s4)
stk.push(s5)

#display
print(stk)

#pop
print(" pop")
stk.pop()
print(stk)

print("pop")
stk.pop()
print(stk)

print(slist)
```
